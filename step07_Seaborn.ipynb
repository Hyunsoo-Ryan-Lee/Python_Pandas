```python
# 레스토랑 팁 데이터 셋 활용
'''
미션 : tip을 어떤 사람이 많이 주는지
'''

tips = sns.load_dataset("tips")
tips
```

```python
tips.columns
```

```python
# 계산한 금액과 tip의 관계도 확인 - 산점도
# 축별 t

sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
# 스타일 적용

sns.set_style(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
# 스타일 적용, feature를 색으로 구분

sns.set_style(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue = 'day')
```

```python
# 스타일 적용, feature를 색으로 구분

sns.set_style(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue = 'size', style='smoker')
```

```python
# 스타일 적용, feature를 색으로 구분

sns.set_style(style='ticks')
sns.relplot(x='total_bill', y='tip', data=tips, hue = 'size', style='smoker', palette='Blues', size='size')
```

```python
tips['day'].value_counts()
```

```python
# 요일별 tip이 수행된 data를 시각화

sns.catplot(x='day', y='tip', data=tips);
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='swarm');
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin');
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='point', hue='day');
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

# Load the example flights dataset and convert to long-form
flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax);
```

```python
flights = flights_long.pivot('month','year','passengers')
flights
```

```python
# Draw a heatmap with the numeric values in each cell

f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax);
```

```python
# annot=True  >>>  각 data 표현, 생략시엔 수치값이 안보임
# f, ax = plt.subplots(figsize=(9, 6))  >>> 생략시엔 최소 사이즈로 그려짐. 사이즈 조절 등은 matplotlib 필요
# linewidths=.5,  >>>  칸들 사이의 여백
# fmt="d"  >>>  정수형으로 표기해주는 포맷

f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax);
```

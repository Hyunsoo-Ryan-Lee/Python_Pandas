```python
area = pd.DataFrame([{'규모구분':'전체', '등급':'A'},{'규모구분':'전용면적 60㎡이하', '등급':'B'},{'규모구분':'전용면적 60㎡초과 85㎡이하', '등급':'C'},
                     {'규모구분':'전용면적 85㎡초과 102㎡이하', '등급':'D'},{'규모구분':'전용면적 102㎡초과', '등급':'E'}])

area
```

```python
df3 = pd.merge(df2,area,on='규모구분')
s1 = df3.groupby(df3['지역명']).mean()
s2 = s1['분양가격(㎡)']
s2
```

```python
font_name = mpl.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family=font_name)
```

```python
ax = s2.plot(kind = 'bar',title='도시별 평균 분양가격',legend=True)
ax.set_xlabel('도시')
ax.set_ylabel('분양가격')
ax.legend(['단위: 천원/㎡'])
```

```python
c1 = df3[df3['지역명']=='서울'].groupby(df3['등급']).mean()
c1 = c1.reset_index()
del c1['연도']
del c1['월']
c2 = c1.set_index('등급')
c2
```
